```python
import pandas as pd
import plotly.graph_objects as go
import plotly.io as pio

# Initialize the radar chart
fig = go.Figure()

# Extract sensor names and criteria
sensors = df.columns[1:]  # Skip the 'Category' column
criteria = df["Category"].tolist()

# Add a trace for each sensor
for sensor in sensors:
    values = df[sensor].tolist()
    # Ensure line closure by repeating the first value at the end
    values.append(values[0])
    closed_criteria = criteria + [criteria[0]]
    
    fig.add_trace(go.Scatterpolar(
        r=values,
        theta=closed_criteria,
        name=sensor,
        mode='lines',  # lines only, no markers
        line=dict(width=4)
    ))

# Update layout with white theme and cleaner look
fig.update_layout(
    template='plotly_white',
    polar=dict(
        radialaxis=dict(
            visible=True,
            range=[1, 5],
            tickvals=[1, 2, 3, 4, 5],
            tickfont=dict(size=10)
        ),
        angularaxis=dict(
            tickfont=dict(size=10)
        )
    ),
    showlegend=True,
    # title="Sensor Comparison Including Event Camera"
)

# Show the figure
fig.show()
# Save as JSON
fig_json = fig.to_json()
with open("../../../static/data/charts/automotive_sensors.json", "w") as f:
    f.write(fig_json)
# pio.write_image(fig, "radar_plot1.png", width=800, height=600)
```

```python
import pandas as pd
import plotly.graph_objects as go
import plotly.io as pio

# Load your data
df = pd.read_csv("sensor_comparison.csv")  # Replace with your actual path if needed

# Initialize the radar chart
fig = go.Figure()

# Extract sensor names and criteria
sensors = df.columns[1:]  # Skip the 'Category' column
criteria = df["Category"].tolist()
closed_criteria = criteria + [criteria[0]]  # Close the loop for radar

# Add a trace for each sensor with styling improvements
for sensor in sensors:
    values = df[sensor].tolist()
    values.append(values[0])  # Close the loop

    fig.add_trace(go.Scatterpolar(
        r=values,
        theta=closed_criteria,
        name=sensor,
        mode='lines+markers',
        line=dict(width=2),
        marker=dict(size=6),
        fill='toself',
        opacity=0.3
    ))

# Update layout with styling and spacing tweaks
fig.update_layout(
    template='plotly_white',
    polar=dict(
        radialaxis=dict(
            visible=True,
            range=[1, 5],
            tickvals=[1, 2, 3, 4, 5],
            tickfont=dict(size=11)
        ),
        angularaxis=dict(
            tickfont=dict(size=11)
        )
    ),
    legend=dict(
        font=dict(size=12),
        orientation="h",
        yanchor="bottom",
        y=-0.2,
        xanchor="center",
        x=0.5
    ),
    margin=dict(l=50, r=50, t=50, b=100),
)

# pio.write_image(fig, "radar_plot2.jpg", width=800, height=600)
# Show the figure
fig.show()
```
